####     The company charges  
####  (i) 25% on the orders having final price (discounted price) greater than 600

#### (ii) 15% on the orders having final price (discounted price) greater than 350 but less than or equal to 600

#### (iii) 10% on the orders having final price (discounted price) greater than 100 but less than or equal to 350

#### (iv) Otherwise, 5% on the final price (discounted price)    </font> 

## Q 5. Find the net revenue generated by the E-Commerce company over all orders placed. (6 marks)

### Ans 5

### ##To find the net revenue generated by the E-Commerce company over all orders placed first we need to get all the discounted price list, for each and every discounted price we need to use if-else condition to apply the E-commerce company charges as per the rule mentioned above.

### We need to use SUM() to get the net revenue over all orders placed.

```python
#Formula to get net revenue generated by the E-Commerce company over all orders placed.
net_revenue_list_ECom = []
discounted_price_list_Ecom = ECom_Data['discounted_price']

for discounted_price in discounted_price_list_Ecom:
    if discounted_price > 600:
        net_revenue = (discounted_price*(25/100))
        net_revenue_list_ECom.append(net_revenue)
    elif discounted_price >350 and discounted_price <=600:
        net_revenue = (discounted_price*(15/100))
        net_revenue_list_ECom.append(net_revenue)
    elif discounted_price >100 and discounted_price <=350:
        net_revenue = (discounted_price*(10/100))
        net_revenue_list_ECom.append(net_revenue)
    else:
        net_revenue = (discounted_price*(5/100))
        net_revenue_list_ECom.append(net_revenue)
```

```python
net_rev_Ecom = sum(net_revenue_list_ECom)
print('Net revenue generated by the E-Commerce company over all orders placed is : ', "{:.2f}".format(net_rev_Ecom))
```

## Q6.Calculate the Net BrandRevenue for each Brand and list the top 14 Brand having maximum revenue in descending order (6 marks)
#### Hint: Net BrandRevenue is BrandRevenue which is generated after all the deductions. Also, BrandRevenue is different from the E-Commerce company's net revenue.

### Ans 6

### ##To find the net  Brand revenue generated by each Brand we need to get all the discounted price list, for each and every discounted price we need to use if-else condition to apply the Brand charges as per the rule mentioned above.

### We need to use GROUPBY() and SUM() to get the net brand revenue for each brand individually.

```python
#Formula to get net brand revenue generated by each individual brand
brand_list = ECom_Data['Brand']
net_Brand_revenue_list_retailer = []
discounted_price_list_brand = ECom_Data['discounted_price']

for discounted_price in discounted_price_list_brand:
    if discounted_price > 600:
        net_revenue = discounted_price-(discounted_price*(25/100))
        net_Brand_revenue_list_retailer.append(net_revenue)
    elif discounted_price >350 and discounted_price <=600:
        net_revenue = discounted_price-(discounted_price*(15/100))
        net_Brand_revenue_list_retailer.append(net_revenue)
    elif discounted_price >100 and discounted_price <=350:
        net_revenue = discounted_price-(discounted_price*(10/100))
        net_Brand_revenue_list_retailer.append(net_revenue)
    else:
        net_revenue = discounted_price-(discounted_price*(5/100))
        net_Brand_revenue_list_retailer.append(net_revenue)
```

```python
#Creating a Data-Frame with index as brand
df_net_brand_rev_info = pd.DataFrame(
    {'Brand': brand_list,
     'Net_brand_revenue': net_Brand_revenue_list_retailer
    })
#To reset the index:
df_net_brand_rev_info = df_net_brand_rev_info.set_index('Brand')
```

```python
#Groupping by brand(s) and getting the total for each brand
df_net_brand_rev_info = df_net_brand_rev_info.groupby(['Brand'])['Net_brand_revenue'].sum()
df_net_brand_rev_info
```

### #The top 14 Brand having maximum revenue :
#### We need to sort values using 'sort_values()' to get top 14 maximum revenue

```python
#Sorting the value and displaying top 14 records.
df_net_brand_rev_info.sort_values(ascending=False).head(14)
```
